```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# line Plot
# Two continuous variables.To see the relationship between the variables.On one of the axis has to be time component 
```

```python
x = [0, 2, 4, 6, 8]
y = [0, 10, 16, 22, 28]
plt.plot(x,y)
plt.show()
```

```python
# Pie Charts
# This is for categorical var 
```

```python
labels = ['Python', 'Pandas', 'Numpy', 'Matplotlib']
sizes = [40, 15, 20, 25]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
# Bar Charts
# To see the count or distribution of the categorical variable 
```

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 6]
plt.bar(categories, values)
plt.show()
```

```python
# Scatter Plots
# two continuous variables.To see the relationship between the variables 
```

```python
x = [1, 2, 3, 4, 5]
y = [5, 7, 9, 11, 13]
plt.scatter(x, y)
plt.show()
```

```python
# Histograms plots
# One continuous variable.To see the distribution of the data
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=30)
plt.show()
```

```python
# Stack Plot
# A stack plot is useful for visualizing multiple data series that contribute to a whole over time. 
# It shows the cumulative contribution of different categories.
```

```python
days = [1, 2, 3, 4, 5]
category1 = [3, 4, 5, 4, 3]
category2 = [2, 2, 3, 4, 5]
category3 = [1, 3, 2, 3, 4]
plt.stackplot(days, category1, category2, category3, labels=['Category 1', 'Category 2', 'Category 3'], colors=['r', 'g', 'b'])
plt.show()
```

```python
# Stem Plot
# A stem plot is used to visualize discrete data points. It is useful for emphasizing individual values in a dataset.
```

```python
x = np.linspace(0, 10, 10)
y = np.sin(x)
plt.stem(x, y, linefmt='r-', markerfmt='bo', basefmt='k-')
plt.show()
```

```python
# Box Plot
# A box plot is useful for summarizing the distribution of data, including median, quartiles, and outliers.
```

```python
data = [np.random.normal(0, std, 100) for std in range(1,5)]
plt.boxplot(data, patch_artist=True)
plt.show()
```

```python
# Error Plot
# An error plot is used to visualize uncertainty in data. It is commonly used in scientific and engineering applications.
```

```python
x = np.linspace(0, 10, 10)
y = np.sin(x)
error = np.random.uniform(0.1, 0.3, size=y.shape)
plt.errorbar(x, y, yerr=error, fmt='o', capsize=5, label='Data with Error')
plt.show()
```

```python
# Violin Plot
# A violin plot is used to visualize the distribution of data while showing density, unlike a box plot.
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 5)]
plt.figure(figsize=(8, 6))
sns.violinplot(data=data)
plt.show()
```

```python
# 3D Plot
# 3D plots are useful for visualizing relationships between three variables. They are commonly used in scientific and engineering applications.
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5, 5, 30)
y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

plt.show()
```
